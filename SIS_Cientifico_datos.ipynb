```python
import pandas as pd
df=pd.read_csv('costos_primas.csv',index_col=0)
print(df.head())
```

```python
df.info()
```

```python
df.A = pd.Categorical(df.A)
df.B = pd.Categorical(df.B)
df.C = pd.Categorical(df.C)
df.D = pd.Categorical(df.D)
df.E = pd.Categorical(df.E)
df.F = pd.Categorical(df.F)
df.G = pd.Categorical(df.G)
df.car_value = pd.Categorical(df.car_value)
df.state = pd.Categorical(df.state)
print(df.info())
```

```python
df['homeowner'] = pd.Categorical(df['homeowner'])
df['married_couple'] = pd.Categorical(df['married_couple'])
df['C_previous'] = pd.Categorical(df['C_previous'])
print(df.info())
```

```python
df_dummies = pd.get_dummies(df)
print(df_dummies.info())
```

```python
from sklearn.model_selection import train_test_split

X=df_dummies.drop(columns=['cost']).copy()
y=df_dummies['cost'].copy()

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1337)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import statsmodels.api as sm
results = sm.OLS(y_train, X_train).fit()
#print(results.summary())
```

```python
print(results.aic)
print(results.rsquared)
```
